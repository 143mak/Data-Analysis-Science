```python
import pandas as pd
import numpy as np
import seaborn as sbn
```

```python
df=pd.read_csv('studentsperformance.csv')
df.head()
```

```python

df.head()
```

```python
df['Total marks']=df['math score'] + df['reading score'] +df['writing score']
df.head()
```

```python
df['maximum marks']=300
df.head()
```

```python
df.describe()
```

```python
df.info(verbose=False)
```

```python
df['percent'] = (df['Total marks']*100000 / df['maximum marks'].sum())
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
sbn.catplot(x="gender", y="percent", kind="bar",
            data=df,aspect=2)
```

```python
sbn.catplot(x="parental level of education", y="percent", kind="point",
            data=df,aspect=2)
```

```python
sbn.catplot(x="race", y="percent", kind="bar",
            data=df,aspect=2)
```
