```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df.head()
```

```python
inputs=df.drop('Survived',axis='columns')
inputs.head()
```

```python
target=df['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass=LabelEncoder()
le_Sex=LabelEncoder()
le_Age=LabelEncoder()
le_Fare=LabelEncoder()

inputs['Pclass_n']=le_Pclass.fit_transform(inputs['Pclass'])
inputs['Sex_n']=le_Pclass.fit_transform(inputs['Sex'])
inputs['Age_n']=le_Pclass.fit_transform(inputs['Age'])
inputs['Fare_n']=le_Pclass.fit_transform(inputs['Fare'])
inputs.head()
```

```python
inputs_n=inputs.drop(['Pclass','Age','Sex','Fare'], axis='columns')
inputs_n.head()
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
inputs_m=inputs_n.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns')
inputs_m.head()
```

```python
model.fit(inputs_m,target)
```

```python
model.score(inputs_m,target)
```

```python
model.predict([[1,0,33,50]])
```
